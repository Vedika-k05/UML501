```python
#PART 1
import pandas as pd

df=pd.read_csv("AWCustomers.csv")
print(df.head())
```

```python
df1=df.drop(columns=['CustomerID','Title','MiddleName','FirstName','LastName','Suffix','AddressLine2','AddressLine1','City','PostalCode','PhoneNumber','LastUpdated',])
from datetime import datetime
#convert to age
df1['BirthDate'] = pd.to_datetime(df1['BirthDate'])
current_year = 2025
df1['Age'] = current_year - df1['BirthDate'].dt.year
df1.drop('BirthDate', axis=1, inplace=True)
df1.head()
```

,TotalChildren,age

```python
#PART2
#a
df1.isnull().sum()
```

```python
#b
df1['YearlyIncomenorm']=(df1['YearlyIncome'] - df1['YearlyIncome'].min()) / (df1['YearlyIncome'].max() - df1['YearlyIncome'].min())
df1[['YearlyIncome','YearlyIncomenorm']].head(10)
```

```python
print(df1.info())
```

```python
#c
df1['YearlyIncomecatg']=pd.qcut(df1['YearlyIncome'],q=3,labels=['Low', 'Mid','High'])
df1[['YearlyIncome','YearlyIncomecatg']].head(10)

```

```python
df1['Agecatg']=pd.cut(df1['Age'],bins=[0,25,55, df1['Age'].max()], labels=['Young', 'Adult','Old'])
df1[['Age', 'Agecatg']].head(10)
```

```python
#d
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
scaled=s.fit_transform(df1[['Age','YearlyIncome']])
# Add new columns to df1
df1['Age_scaled']=scaled[:, 0]
df1['YearlyIncome_scaled']=scaled[:, 1]

df1.head()

```

```python
#e
cols=['Gender','MaritalStatus','HomeOwnerFlag','Education','Occupation']#nominal attributes
# One-Hot Encoding
df1_enc=pd.get_dummies(df1, columns=cols, drop_first=True)
print(df1_enc.head())
```

```python
#PART3
import numpy as np

binary_cols=[col for col in df1_enc.columns if set(df1_enc[col].unique()).issubset({0,1})]
obj1=df1_enc[binary_cols].iloc[0].values #value for arr conversion
obj2=df1_enc[binary_cols].iloc[1].values
# Simple Matching
match=np.sum(obj1 == obj2)
smc = match/len(obj1)
print(smc)

```

```python
# Jaccard Similarity

intersection=np.sum((obj1 == 1)&(obj2 == 1))
union=np.sum((obj1 == 1)|(obj2 == 1))
jaccard=intersection/union 
print("Jaccard",jaccard)
```
